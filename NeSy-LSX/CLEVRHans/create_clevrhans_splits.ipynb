```python
import os
import json
import numpy as np
```

```python
base_path = "/pathto/CLEVR-Hans3/"
split = "train"

scenes_path = os.path.join(
    base_path, split, "CLEVR_HANS_scenes_{}.json".format(split)
)

with open(scenes_path) as fd:
    scenes = json.load(fd)["scenes"]
```

```python
np.random.permutation(scenes)
```

```python
n_scenes_val_lsx = 1500
n_scenes_train_lsx = len(scenes) - n_scenes_val_lsx

scenes_train_lsx = scenes[:n_scenes_train_lsx]
scenes_val_lsx = scenes[n_scenes_train_lsx:]
```

```python
def write_ver_json(base_path, split, scenes, ver):
    fp = os.path.join(
        base_path, split, f"CLEVR_HANS_scenes_{split}_{ver}.json"
    )
    with open(fp, 'w') as outfile:
        tmp_dict = {'scenes': scenes,
                    'info': f"CLEVR Hans 3 version {ver}"}
        json.dump(tmp_dict, outfile)
    
write_ver_json(base_path, split, scenes_train_lsx, 'train_lsx')
write_ver_json(base_path, split, scenes_val_lsx, 'val_lsx')
```
